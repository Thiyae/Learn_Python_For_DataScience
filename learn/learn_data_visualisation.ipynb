```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
os.getcwd()
```

```python
cars1=pd.read_csv('C:\\Users\\thiya\\learn/updated_cars.csv',index_col=0)
```

```python
cars=cars1.copy()
```

```python
cars
```

```python
cars.isnull().sum()
```

```python
cars.dropna(axis=0,inplace=True)
```

```python
cars.isnull().sum()
```

```python
cars.info()
```

```python
cars.KM.replace("??",np.nan,inplace=True)
cars.KM.replace("????",np.nan,inplace=True)
cars.KM=cars.KM.astype('float')
```

**SCATTER PLOT**

scatter plot is mostly used to get interpret the correlation between variables

```python
plt.scatter(cars.Age,cars.Price,c='red')
plt.title('Scatter plot of Price vs Age of the cars')
plt.xlabel('Age (months)')
plt.ylabel('Price (Euros)')
plt.show()
```

**HISTOGRAM**

Histogram often used to get the frequency distribution of numerical varibles

```python
plt.hist(cars.KM)
```

```python
plt.hist(cars.KM,color='green',edgecolor='white',bins=5)
plt.title('Histogram of Kilometer')
plt.xlabel('Kilometer')
plt.ylabel('Frequency')
plt.show()
```

**BAR PLOT**

To represent the frequency distribution of categorical variables

A bar diagram makes it easy to compare sets of data between different groups

```python
cars.FuelType.value_counts()
```

```python
counts=cars.FuelType.value_counts()
fuelType=('Petro','Diesel','CNG')
index=np.arange(len(fuelType))
counts
```

```python
plt.bar(index,counts,color=['red','blue','cyan'])
plt.title('Bar plot of fuel types')
plt.xlabel('Fuel Types')
plt.ylabel('Frequency')
plt.xticks(index,fuelType,rotation=90)
plt.show()
```
