```python
#joint of probability of two independent events
pd.crosstab(index=cars.Automatic,columns=cars.FuelType,normalize=True,dropna=True)
```

```python
#margin probability of a single event
pd.crosstab(index=cars.Automatic,columns=cars.FuelType,margins=True,normalize=True,dropna=True)
```

```python
#conditional for two independent varibales
pd.crosstab(index=cars.Automatic,columns=cars.FuelType,normalize='index',dropna=True)
```
