```python
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
os.getcwd()
```

```python
cars=pd.read_csv("new_cars_data.csv",index_col=0)
```

```python
cars.KM.replace("??",np.nan)
cars['KM']=cars.KM.astype('float')
```

```python
cars.isnull().sum()
```

```python
missing_rows=cars[cars.isnull().any(axis=1)]
```

```python
missing_rows
```

```python
missing_rows.info()
```

```python
cars.describe()
```

```python
missing_rows.Age=missing_rows.Age.fillna(cars.Age.mean())
```

```python
missing_rows
```

```python
missing_rows.info()
```

```python
missing_rows.KM=missing_rows.KM.fillna(cars.KM.median())
```

```python
missing_rows.HP=missing_rows.HP.fillna(cars.HP.mean())
```

```python
missing_rows.FuelType=missing_rows.FuelType.fillna(cars.FuelType.value_counts().index[0])
```

```python
missing_rows
```

```python
missing_rows.MetColor=missing_rows.MetColor.fillna(cars.FuelType.mode()[0])
```
